```python
#visualization with matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create arrays
x = np.linspace(0,10,)
```

```python
#create compute function
y = x**10 + 4
```

```python
plt.plot(x,y,label = 'y=x**10 + 4',color ='red',linestyle = '-')
```

```python
plt.plot(x,y,label = 'y=x**10 + 4',color ='red',linestyle = '-')
plt.xlabel('x axis')
plt.ylabel('y axix')
plt.title('Quadratic function')
plt.grid(True)
```
