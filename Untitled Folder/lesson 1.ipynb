```python
#visualization with matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create arrays
x = np.linspace(0,10,)
```

```python
#create compute function
y = x**8 + 10*x + 4
```

```python
#plotting
plt.plot(x,y,label = 'y=x**8 + 10*x + 4',color ='red',linestyle = '-')
```
