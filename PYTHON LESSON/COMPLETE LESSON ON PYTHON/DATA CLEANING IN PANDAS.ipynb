```python
df = pd.read_excel(r"C:\Users\USER\Desktop\ML\Customer Call List.xlsx")
```

```python
df.drop(columns='Not_Useful_Column',inplace = True)
```

```python
df['Last_Name']=df['Last_Name'].str.strip('123._/')
df
```

```python
import re
df['Phone_Number'] = df['Phone_Number'].apply(
    lambda phone: None if pd.isna(phone)
    else re.sub(r'[\-|/|]', '', phone)
)
```

```python
df
```

```python
# Replace 'NaN', 'Na', and 'None' in Phone_Number column with None
df['Phone_Number'] = df['Phone_Number'].replace(['NaN', 'Na', None, 'N/a'], None)
```

```python
df['Paying Customer'] = df['Paying Customer'].apply(
    lambda value: 'No' if pd.isna(value) or value.lower() == 'n' else ('Yes' if value.lower() in ['yes', 'y'] else value)
)
```
